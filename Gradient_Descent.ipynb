# What is Gradient Descent?
Gradient descent is an optimization algorithm used to minimize the cost function in machine learning. In simpler terms, it's a method for finding the best values of the parameters in a model that will make the model perform as well as possible.

# How it works?
First we will take the derivative of the loss function for each perameter in it.
Then we will pick a random value for the perameter and plug the perameter value in the derivative.
After that we will calculate the step size i.e. Step Size = Slope x Learning Rate.
Then we will calculate the new peramater i.e new perameter = old perameter - step size.
Then we will place the value of the new perameter in the derivative.
We repeat this process from step 2 until the step size get really small or you reach the maximum number of steps.

```python
import numpy as np

X = [0.5, 2.5]
Y = [0.2, 0.9]
```

Here we have made a sigmoid with perammeters w, x, and b

The formula of sigmoid is y = 1 / 1 + e**(-(w*x + b))

```python
def f(w,x,b):
  return 1.0/(1.0 + np.exp(-(w*x + b)))
```

Now, we are going to make a method to find the error.
Using the formula given below we will find the error.
We want L(w,b) to be as close to 0 as possible.

image.png

```python
def error(w,b):
  err = 0.0
  for x,y in zip(X,Y):
    fx = f(w,x,b)
    err += 0.5 * (fx - y)**2
  return err
```

Here we are finding the gradient of weight and biase.
first we will find the derivative of value L(w,b) with respect to w and we will left with the formula (fx - y) * fx * (1-fx) * x which is the gradient of weight or dell weight and we will do the same for the biase we will find the derivative of value L(w,b) with respect to b and the end product is (fx - y) * fx * (1-fx) which is the gradient of biase.

```python
def grad_w(w,b,x,y):
  fx = f(w,b,x) 
  return (fx - y) * fx * (1-fx) * x

def grad_b(w,b,x,y):
  fx = f(w,b,x)
  return (fx - y) * fx * (1-fx)
```

At last, we will write the code to train the model.
And at the end in the results we can see the model finding the gradient descent. 

```python
def do_gradient_descent():
  w,b,learning_rate,max_epochs = 0,0,0.1,100
  for i in range(max_epochs):
    dw,db=0,0
    for x,y in zip(X,Y):
      dw += grad_w(w,b,x,y)
      db += grad_b(w,b,x,y)
    w = w - learning_rate * dw
    b = b - learning_rate * db
    print(w,b)
    print(error(w,b))
```

```python
do_gradient_descent()
```
